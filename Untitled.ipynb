## Import

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, classification_report
from scipy.sparse import hstack
```

## Load Data

```python
df = pd.read_csv("Resume.csv")

df = df[["Resume_str", "Category"]]
df = df.dropna()

df["Category"] = df["Category"].str.lower().str.strip()

print(df["Category"].value_counts())
```

## Cleaning Functions

```python
def clean_resume(text):
    text = re.sub(r'<.*?>', ' ', text)
    text = re.sub(r'http\S+|www\S+', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'[^a-zA-Z ]', ' ', text)
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]
    return " ".join(words)

def advanced_clean(text):
    text = clean_resume(text)
    text = remove_stopwords(text)
    words = text.split()
    words = [w for w in words if 3 <= len(w) <= 15]
    return " ".join(words)
```

```python
df["cleaned"] = df["Resume_str"].apply(advanced_clean)
```

```python
group_map = {
    "information-technology": "tech",
    "engineering": "tech",
    "finance": "finance",
    "accountant": "finance",
    "banking": "finance",
    "hr": "management",
    "business-development": "management",
    "consultant": "management",
    "sales": "sales",
    "public-relations": "sales",
    "healthcare": "healthcare",
    "designer": "creative",
    "arts": "creative",
    "digital-media": "creative",
    "apparel": "creative",
    "teacher": "education",
    "chef": "hospitality",
    "aviation": "aviation",
    "construction": "construction",
    "fitness": "fitness",
    "agriculture": "agriculture",
    "automobile": "automobile",
    "bpo": "bpo"
}

df["Category"] = df["Category"].map(group_map)
df = df.dropna()
```

```python
le = LabelEncoder()
df["label"] = le.fit_transform(df["Category"])

label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(label_mapping)
```

## Keyword Boost

```python
def keyword_boost(text):
    tech_keywords = ["python", "react", "node", "machine learning"]
    score = 0
    
    for word in tech_keywords:
        if word in text.lower():
            score += 0.05
            
    return score

```

## LIME (EXPLAINABILITY)

```python
from lime.lime_text import LimeTextExplainer

class_names = list(le.classes_)
explainer = LimeTextExplainer(class_names=class_names)

def predict_proba_lime(texts):
    cleaned = [advanced_clean(t) for t in texts]
    
    word_vec = vectorizer.transform(cleaned)
    char_vec = char_vectorizer.transform(cleaned)
    
    vec = hstack([word_vec, char_vec])
    
    return model.predict_proba(vec)

def explain_prediction(text):
    exp = explainer.explain_instance(
        text,
        predict_proba_lime,
        num_features=15
    )
    
    # Filter weak/noisy words
    return [(w, s) for w, s in exp.as_list() if abs(s) > 0.01]
```
